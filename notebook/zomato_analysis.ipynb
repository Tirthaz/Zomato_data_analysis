```python
df= pd.merge(
    zomato_df,
    country_df,
    how="left",
    on="Country Code"
)
```

```python
df.head()
```

```python
df["Aggregate rating"] = pd.to_numeric(
    df["Aggregate rating"], errors="coerce"
)

df = df[df["Aggregate rating"] > 0]
```

```python
city_count = df["City"].value_counts().head(10)

plt.figure(figsize=(10,5))
city_count.plot(kind="bar")
plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
country_count = df["Country"].value_counts().head(10)

plt.figure(figsize=(10,5))
country_count.plot(kind="bar")
plt.title("Top Countries with Restaurants on Zomato")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

```python
delivery_rating = df.groupby("Has Online delivery")["Aggregate rating"].mean()

delivery_rating
```

```python
delivery_rating.plot(kind="bar")
plt.title("Online Delivery vs Average Rating")
plt.ylabel("Average Rating")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    x="Price range",
    y="Aggregate rating",
    data=df
)
plt.title("Price Range vs Rating")
plt.show()
```

```python
from collections import Counter

cuisines = df["Cuisines"].dropna().str.split(", ")
cuisine_list = [c for sublist in cuisines for c in sublist]

cuisine_count = Counter(cuisine_list).most_common(10)
cuisine_df = pd.DataFrame(cuisine_count, columns=["Cuisine", "Count"])

cuisine_df
```

```python
plt.figure(figsize=(10,5))
plt.bar(cuisine_df["Cuisine"], cuisine_df["Count"])
plt.xticks(rotation=45)
plt.title("Top 10 Most Popular Cuisines")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(
    df["Average Cost for two"],
    df["Aggregate rating"],
    alpha=0.5
)
plt.xlabel("Average Cost for Two")
plt.ylabel("Rating")
plt.title("Cost vs Rating")
plt.show()
```

1.India has the highest number of Zomato-listed restaurants

2.Mid-priced restaurants have better ratings than very cheap or expensive ones

3.Online delivery restaurants tend to have slightly higher ratings

4.North Indian & Chinese cuisines dominate
